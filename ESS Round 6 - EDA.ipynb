## Data Cleaning
### 1. Encoding 1-5 scale items
Reverse wording items

```python
#Encoding 1-5 scale items
#step 1: scale 1-5 : agree strongly - disagree strongly
ess_df_select.loc[:,['dclvlf', 'lchshcp', 'accdng','dngval','optftr',
    'lotsgot', 'pstvms', 'flrms','wrbknrm','flclpla']] = ess_df_select[[ 
    'dclvlf', 'lchshcp', 'accdng','dngval','optftr',
    'lotsgot', 'pstvms', 'flrms','wrbknrm','flclpla']].replace([1,2,4,5],['A','B','D','E'], regex= True)
```

```python
#Encoding 1-5 scale items
#step 2: scale 1-5 : disagree strongly - agree strongly
ess_df_select[[ 
    'dclvlf', 'lchshcp', 'accdng','dngval','optftr',
    'lotsgot', 'pstvms', 'flrms','wrbknrm','flclpla']] = ess_df_select[[ 
    'dclvlf', 'lchshcp', 'accdng','dngval','optftr',
    'lotsgot', 'pstvms', 'flrms','wrbknrm','flclpla']].replace(['A','B','D','E'], [5,4,2,1], regex= True)
```

### 2. Reverse value items

```python
ess_df_select[[
    'fltsd','fltdpr','fltanx','lchshcp','flrms',
'wrbknrm','flteeff','slprl','cldgng','fltlnl']] = ess_df_select[[
    'fltsd','fltdpr','fltanx','lchshcp','flrms',
'wrbknrm','flteeff','slprl','cldgng','fltlnl']] * -1

ess_df_select[[
    'fltsd','fltdpr','fltanx','lchshcp','flrms',
'wrbknrm','flteeff','slprl','cldgng','fltlnl']].head()
```

### 3. Missing values

```python
# cldgng: dtype object  => country AL doesnt answer this item
#Fill no answer cells of country AL in item "cldgng" to 0
ess_df_select.loc[:,['cldgng']] = ess_df_select['cldgng'].replace(r'^\s*$',0, regex= True)

#Convert "cldgng" from object to number
ess_df_select.loc[:,['cldgng']]= pd.to_numeric(ess_df_select['cldgng'])

ess_df_select.loc[:,[
    'fltsd','fltdpr','fltanx','lchshcp','flrms',
'wrbknrm','flteeff','slprl','cldgng','fltlnl']].head()
```

```python
#Identify missing values
# items with scale 0-10: replace 99,88,77 => NaN

ess_df_select = ess_df_select.replace([99,88,77],np.nan, regex= True)

# other items: replace 9,8,7 => NaN
ess_df_select[['fltsd','fltdpr','enjlf','wrhpp','fltanx','fltpcfl','dclvlf','lchshcp',
'accdng','dngval','optftr','lotsgot','pstvms','flrms','wrbknrm','flteeff','slprl',
'cldgng','enrglot','pplahlp','flclpla','inprdsc','rehlppl','fltlnl']] = ess_df_select[['fltsd','fltdpr','enjlf','wrhpp','fltanx','fltpcfl','dclvlf','lchshcp',
'accdng','dngval','optftr','lotsgot','pstvms','flrms','wrbknrm','flteeff','slprl',
'cldgng','enrglot','pplahlp','flclpla','inprdsc','rehlppl','fltlnl']].replace([7,8,9, -7, -8, -9],np.nan, regex= True)

#Drop NaN
ess_df_select.dropna(inplace=True)

ess_df_select.info()

```
