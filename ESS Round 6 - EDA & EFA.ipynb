## Data Cleaning
### 1. Encoding 1-5 scale items
Reverse wording items

```python
#Encoding 1-5 scale items
#step 1: scale 1-5 : agree strongly - disagree strongly
df_15scale = ess_df_1[['dclvlf','lchshcp', 'accdng','dngval','optftr',
    'lotsgot', 'pstvms', 'flrms','wrbknrm','flclpla']].copy()
df_15scale
```

```python
df_15scale = df_15scale.replace([1,2,4,5],['A','B','D','E'])
```

```python
#Encoding 1-5 scale items
#step 2: scale 1-5 : disagree strongly - agree strongly
df_15scale = df_15scale.replace(['A','B','D','E'], [5,4,2,1])
df_15scale
```

```python
#Create a new DF

#df without 1-5 scale items
ess_df_2= ess_df[['stflife','happy','fltsd','fltdpr','enjlf','wrhpp','fltanx',
'fltpcfl','tmimdng','tmabdng','tmendng','sedirlf','deaimpp','flteeff','slprl',
'cldgng','enrglot','ppltrst','pplfair','pplhlp','pplahlp','inprdsc','flapppl','rehlppl','fltlnl']]
#merge with 1-5scale items
ess_df_select= pd.concat([ess_df_2,df_15scale], axis=1)
ess_df_select.shape
```

### 3. Missing values

```python
# cldgng: dtype object  => country AL doesnt answer this item
#Fill no answer cells of country AL in item "cldgng" to 0
ess_df_select['cldgng'] = ess_df_select['cldgng'].replace(r'^\s*$',0, regex= True)

#Convert "cldgng" from object to number
ess_df_select['cldgng']= pd.to_numeric(ess_df_select['cldgng'])
ess_df_select['cldgng']
```

```python
#Identify missing values
# items with scale 0-10: replace 99,88,77 => NaN

ess_df_select = ess_df_select.replace([99,88,77],np.nan, regex= True)

# other items: replace 9,8,7 => NaN
ess_df_select[['fltsd','fltdpr','enjlf','wrhpp','fltanx','fltpcfl','dclvlf','lchshcp',
'accdng','dngval','optftr','lotsgot','pstvms','flrms','wrbknrm','flteeff','slprl',
'cldgng','enrglot','pplahlp','flclpla','inprdsc','rehlppl','fltlnl']] = ess_df_select[['fltsd','fltdpr','enjlf','wrhpp','fltanx','fltpcfl','dclvlf','lchshcp',
'accdng','dngval','optftr','lotsgot','pstvms','flrms','wrbknrm','flteeff','slprl',
'cldgng','enrglot','pplahlp','flclpla','inprdsc','rehlppl','fltlnl']].replace([7,8,9, -7, -8, -9],np.nan, regex= True)

#Drop NaN
ess_df_select.dropna(inplace=True)

ess_df_select.describe()
```

```python
import pandas as pd
from sklearn.datasets import load_iris
from factor_analyzer import FactorAnalyzer
```

```python
#Load data
EFA_df = pd.read_csv('/Users/ninh/Desktop/EES_well-being-clean data.csv')

EFA_df.drop(['cntry'],axis=1,inplace=True)
```

```python
# Adequacy Test
# 1. Bartlett’s Test

from factor_analyzer.factor_analyzer import calculate_bartlett_sphericity
chi_square_value,p_value = calculate_bartlett_sphericity(EFA_df)
chi_square_value, p_value
```

```python
from factor_analyzer.factor_analyzer import calculate_kmo
kmo_all,kmo_model=calculate_kmo(EFA_df)
kmo_model
```

### EFA 1

```python
# Create factor analysis object and perform factor analysis
# I. Define the number of factors
fa = FactorAnalyzer(10, rotation=None)
fa.fit(EFA_df)
FactorAnalyzer(bounds=(0.005, 1), impute='median', is_corr_matrix=False,
method='principal', n_factors=3, rotation= 'promax', rotation_kwargs={},
use_smc=True)

ev,v = fa.get_eigenvalues()
ev

# Array 1.The original eigen values
# Array 2.The common factor eigen values
```

```python
ev.shape
```

```python
# SCREEPLOT
plt.scatter(range(1,36),ev)
plt.plot(range(1,36),ev)
plt.title('Scree Plot')
plt.xlabel('Factors')
plt.ylabel('Eigenvalue')
plt.grid()
plt.show()
```

Result: 5 or 7 factors (eigen values >= 1 )

```python
fa.get_factor_variance()

# Total Variance Explained
# Array 1. Sum of squared loadings (variance)
# Array 2. Proportional variance 
# Array 3. Cumulative variance : 47.5%
```

```python
# Factor loading analysis with 7 factors

fa = FactorAnalyzer(7, rotation='promax', method='principal',use_smc=True)
fa.fit(EFA_df)
FactorAnalyzer(bounds=(0.005, 1), impute='median', is_corr_matrix=False,
method='principal', n_factors= 7, rotation= 'promax', rotation_kwargs={},
use_smc=True)
```

```python
# fa.loadings_ : array of all factor loadings of items

loadings = pd.DataFrame(fa.loadings_, columns=['Factor 1', 'Factor 2', 'Factor 3', 'Factor 4','Factor 5','Factor 6','Factor 7'], index=EFA_df.columns)
```

```python
# Present the loading tables
#Color the cell
def color_green(val):
    color = 'green' if np.abs(val) > 0.4 else 'black'
    return 'color: %s' % color
loadings[['Factor 1','Factor 2','Factor 3','Factor 4','Factor 5','Factor 6','Factor 7']].style.applymap(color_green)
```

Result: Loadings of flrms, deaimpp, sedirlf, wrbknrm are low
